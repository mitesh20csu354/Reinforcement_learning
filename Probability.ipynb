```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("student-mat.csv")
```

```python
df.head(5)
```

```python
len(df)
```

```python
df.shape
```

```python
df['grade_A']=np.where(df['G3']*5>=80,1,0)
```

```python
df['high_absenses']=np.where(df['absences']>=10,1,0)
```

```python
df['count']=1
```

```python
df
```

```python
df=df[['grade_A','high_absenses','count']]
df.head()
```

```python
pd.pivot_table(
    df,
    values='count',
    index=['grade_A'],
    columns=['high_absenses'],
    aggfunc=np.size,
    fill_value=0
)
```
